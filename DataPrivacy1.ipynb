```python
##--------------HR(SAME) dataset---------------
# Explore the dataset
df_employees.head()

# Explore the histogram of the age variable
df_employees['Age'].hist(bins=15)
```

```python
##----------Generalization------------
# Apply generalization by transforming to binary data
df_employees['Age'] = df_employees['Age'].apply(lambda x:">=40"if x>=40 else"<40" )

# See results
df_employees.head()
```

```python
##----------Top and bottom coding-----------------
# Explore the histogram of the age variable
df_employees['Age'].hist(bins=15)
```

```python
##---------------Top coding--------------------
# Filter to see rows affected
df_employees[df_employees['Age'] >= '55']

# Top code the age to 55
df_employees.loc[df_employees['Age'] > '55', 'Age'] = '55'

# Filter to see rows affected
df_employees[df_employees['Age'] >= '55']
```

```python
##------Bottom coding-------
# Explore the histogram of the age variable
df_employees['Age'].hist(bins=15)
#Bottom code the age to 25
df_employees.loc[df_employees['Age'] < '25', 'Age'] = '25'
# Explore the histogram of the age variable
df_employees['Age'].hist(bins=15)
```
